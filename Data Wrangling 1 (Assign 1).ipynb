```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# **Importing Movies Dataset**

```python
data = pd.read_csv('movies_data.csv')
print("Number of rows : ", data.shape[0])
print("Number of columns : ", data.shape[1])
print("top 10 sample data : ")
data.head(10)
```

# **Sampling data for 20 values**

```python
sample_data = data.sample(n = 20, random_state = 100)
```

```python
print("Number of rows : ", sample_data.shape[0])
print("Number of columns : ", sample_data.shape[1])
sample_data
```

## **NULL or Missing values count**

```python
sample_data.isnull().sum()
```

## **Descriptive Statistics**

```python
sample_data.describe()
```

## **Changing Column Names**

```python
print(sample_data.columns)
```

```python
sample_data.columns = [column.title() for column in sample_data.columns]
```

```python
sample_data.columns
```

## **Dropping unnecessary columns**

```python
columns_to_delete = ['One-Line', 'Stars', 'Gross', 'Year']
sample_data.drop(columns = columns_to_delete, inplace = True)
```

```python
sample_data
```

## **Show rows with NULL values in any column**

```python
sample_data[sample_data.isnull().any(axis=1)]
```

```python
sample_data.reset_index(drop = True, inplace = True)
```

```python
sample_data
```

## **Filling NULL with value 0**

```python
sample_data.fillna(0, inplace = True)
```

```python
sample_data['Votes'] = sample_data['Votes'].str.replace(',', '').astype(float)
```

```python
sample_data.fillna(0, inplace = True)
sample_data
```
