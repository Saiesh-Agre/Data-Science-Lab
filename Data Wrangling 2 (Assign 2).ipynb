```python
import pandas as pd
import numpy as np

# Create sample data
data = {
    'StudentID': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110,
                  111, 112, 113, 114, 115, 116, 117, 118, 119, 120],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Heidi', 'Ivy', 'Jack',
             'Katie', 'Liam', 'Mia', 'Nora', 'Oliver', 'Penelope', 'Quinn', 'Rose', 'Sam', 'Tom'],
    'Math': [85, 92, 78, 67, np.nan, 88, 90, 79, 82, 95,
             70, 85, 90, 88, 75, 82, 94, 65, 80, 78],
    'Science': [90, 88, 72, 85, 78, 95, np.nan, 83, 88, 91,
                83, 75, 85, 72, 92, 80, 85, 79, 83, 89],
    'English': [80, 85, 92, 78, 75, 90, 88, 93, 82, 87,
                85, 80, 78, 94, 86, 92, np.nan, 85, 90, 95]
}

# Create DataFrame
df = pd.DataFrame(data)

# Introduce inconsistencies, NULL values, and outliers
df.loc[10, 'Math'] = np.nan  # Introducing NULL value
df.loc[15, 'Science'] = np.nan  # Introducing NULL value
df.loc[3, 'Math'] = 150  # Introducing outlier value
df.loc[8, 'Science'] = 200  # Introducing outlier value
df.loc[19, 'English'] = 105  # Introducing outlier value
```

## Filling missing values with mean

```python
df['Math'] = df['Math'].fillna(df['Math'].mean())
df['Science'] = df['Science'].fillna(df['Science'].mean())
df['English'] = df['English'].fillna(df['English'].mean())
```
